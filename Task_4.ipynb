```python
# problem 1
def recharge():
  print("********VALID PLANS*******")
  print("1.Rs.199, 2.Rs.299, 3.Rs.399, 4.Rs.599")
  n=int(input("Enter the plan:"))
  rrecharge=[199,299,399,599]
  if n<50:
    print("amount invalid")
    recharge()
  while n in rrecharge:
    print("recharge succesfull")
    break
  else:
    print("plan not valid")
    recharge()
recharge()
```

```python
#problem 2
storage={"football":25,"cricket ball":10, "batt":15}
for product,stock in storage.items():
  if stock<15:
    print(f"{product} has stock {stock}, Restock alert!!")
  else:
    print(f"{product} has stock {stock}, stock ok")
```

```python
#problem 3
def cal():
  n=input("enter the name:")
  dic={}
  m=[]
  for i in range(3):
    mark=int(input("enter the marks:"))
    m.append(mark)
  dic[n]=m
  for n,m in dic.items():
    avg=sum(m)/3
  print("The average marks is:",avg)
  if avg>=50:
    print("pass")
  else:
    print("fail")
cal()
```

```python
#problem 4
def cal():
  while True:
    n=int(input("Enter the kilometer travelled:"))
    fare=50+(n-1)*12
    x=input("was it peak hours?(y/n):")
    if x=="y":
      f=fare*0.25
      new=fare+f
      print("Cab fare:",new)
    else:
      print("Cabe fare:",fare)
    a=input("do you want to continue?(y/n):")
    if a!="y":
      break
cal()
```

```python
#problem 5
def att():
  x=input("Enter the name of the employee:")
  f=[]
  for i in range(6):
    p=input("enter the attendance for 6 days:")
    f.append(p)
  dic={}
  f.append(p)
  dic[x]=f
  for x,f in dic.items():
    count=0
    for i in f:
      if i=="P":
        count+=1
  avg=(count/6)*100
  if avg>=75:
    print("Eligible")
  else:
    print("not eligible")
att()
```

```python
#problem 6
def check():
    p = input("Enter the password: ")
    if len(p) < 8:
        print("Password Weak")
    else:
      d=0
      l=0
      s=0
      for i in p:
        if i.isdigit():
          d+=1
        elif i.isalpha():
          l+=1
        elif not i.isalnum():
          s+=1
      if d!=0 and l!=0 and s !=0:
          print("Password Strong")
      else:
          print("Password Weak")
check()
```
