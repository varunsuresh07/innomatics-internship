```python
#problem 1
class statuss:
  def __init__(self,name):
    self.name=name
    self.status="OFF"
  def on(self):
    self.status="ON"
  def off(self):
    self.status="off"
  def shows(self):
    print(self.name,"is",self.status)
LightName="Bedroom Light"
Action="ON"
self=statuss(LightName)
if Action=="ON":
  self.on()
else:
  self.off()
self.shows()
```

```python
#problem 2
class employee:
  def __init__(self,name,id,department):
    self.name=name
    self.id=id
    self.department=department
  def showss(self):
    print("Employee Name:",self.name)
    print("Employee ID:",self.id)
    print("Department:",self.department)
EName="Rahul"
EID= "EMP102"
Department="IT"
selff=employee(EName,EID,Department)
selff.showss()
```
